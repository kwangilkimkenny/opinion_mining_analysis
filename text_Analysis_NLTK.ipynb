```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
from pandas import DataFrame as df
from tqdm import tqdm

read_data = '파일네임입력.txt'
with open(read_data) as files:
    contents = files.read()


list_str = contents.split(".")  #문장별로 분리한다. 분리는 .를 기준으로 한다.   


cols = ['compound', 'neg', 'neu', 'pos']
df = pd.DataFrame(index = range(len(list_str)), columns=cols)


sid = SentimentIntensityAnalyzer()

for sentence in tqdm(list_str): #한문장식 가져와서 처리한다
    ss = sid.polarity_scores(sentence) #긍정, 부정, 중립, 혼합점수 계산
    for k in ss:
        #print('{0}: {1}, ' .format(k, ss[k]), end='')
        #print()
        #으로 4번을 돌아가면서 연산하여 출력 긍정, 중립, 부정, 합 값
        #DataFrame에 넣어서 Chart생성데이터 처리
        if k == 'compound': 
            #print('compound')
            #print('{0}: {1}, ' .format(k, ss[k]), end='')
            CompoundValue = int(ss[k])
            type(CompoundValue)
            df_result = df.append(pd.DataFrame([[CompoundValue]], columns=['compound']))
        if k == 'neg':
            #print('neg')
            df_result = df.append(pd.DataFrame([[ss[k]]], columns=['neg']))
        if k == 'neu':
            #print('neu')
            df_result = df.append(pd.DataFrame([[ss[k]]], columns=['neu']))
        if k == 'pos':
            #print('pos')
            df_result = df.append(pd.DataFrame([[ss[k]]], columns=['pos']))
            
        
df_result

```

```python
list_str = contents.split(".")  #문장별로 분리한다. 분리는 .를 기준으로 한다.   

print(list_str[:100])
```

```python
listSentiment = []

sid = SentimentIntensityAnalyzer()


i=0
for sentence in tqdm(list_str): #한문장식 가져와서 처리한다
#     print("sentence:",sentence)
    ss = sid.polarity_scores(sentence) #긍정, 부정, 중립, 혼합점수 계산
#     print(type(ss))
    print(ss.keys())
    print('{}: neg:{},neu:{},pos:{},compound:{}'.format(i,ss['neg'],ss['neu'],ss['pos'],ss['compound']))
    i +=1
    listSentiment.append([ss['neg'],ss['neu'],ss['pos']])

print()
print(listSentiment)
```

```python
import numpy as np

# (['neg', 'neu', 'pos', 'compound'])
# [0.262, 0.738, 0.0, -0.4939]
# neg = -1 , neu = 0  , pos = 1 
# max:{} 0

maxlistSentment = []

for sentiment in listSentiment :
#     print(sentiment)
#     print("max:{}",np.argmax(sentiment)-1)
    
    maxlistSentment.append(np.argmax(sentiment)-1)
    
print(maxlistSentment) 
print(len(maxlistSentment))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,10))
plt.plot(maxlistSentment)
plt.axis([0,len(maxlistSentment),-1,1])
plt.ylabel('sentiment essay')
plt.xlabel('sentences')
plt.show()
```

Naïve의 Bayes 분류기를 사용한 감성분석을 할 수도 있음. 아래는 예제임

```python
import nltk
from nltk.tokenize import word_tokenize
  
# Step 1 – Training data
train = [("Great place to be when you are in Bangalore.", "pos"),
  ("The place was being renovated when I visited so the seating was limited.", "neg"),
  ("Loved the ambience, loved the food", "pos"),
  ("The food is delicious but not over the top.", "neg"),
  ("Service - Little slow, probably because too many people.", "neg"),
  ("The place is not easy to locate", "neg"),
  ("Mushroom fried rice was spicy", "pos"),
]
  
# Step 2 
dictionary = set(word.lower() for passage in train for word in word_tokenize(passage[0]))
  
# Step 3
t = [({word: (word in word_tokenize(x[0])) for word in dictionary}, x[1]) for x in train]
  
# Step 4 – the classifier is trained with sample data
classifier = nltk.NaiveBayesClassifier.train(t)
  
test_data = "Manchurian was hot and spicy"
test_data_features = {word.lower(): (word in word_tokenize(test_data.lower())) for word in dictionary}
  
print (classifier.classify(test_data_features))
```
